```python
import pandas as pd
```

```python
data = {
    'Name': [
        'Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen',
        'Ian', 'Jane',
        'Karl', 'Laura', 'Mike', 'Nina', 'Oscar', 'Paula', 'Quentin', 'Rachel',
        'Steve', 'Tina',
        'Uma', 'Victor', 'Wendy', 'Xander', 'Yara', 'Zane'
    ],
    'Department': [
        'HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'Marketing', 'Sales'
        ,'HR','IT', 'Finance', 'Sales', 'Marketing', 'IT', 'HR', 'Finance', 'Sales',
        'IT', 'Marketing',
        'HR', 'Finance', 'Sales', 'IT', 'Marketing', 'Finance'
    ],
    'Salary': [
        50000, 60000, 55000, 70000, 48000, 62000, 65000, 52000, 58000, 51000,
        75000, 53000, 60000, 57000, 67000, 49500, 61000, 59000, 72000, 54000,
        47000, 56000, 61000, 68000, 55000, 64000
    ],
    'Join_Date': [
        '2020-05-21', '2019-03-15', '2021-07-10', '2018-11-01', '2022-01-05',
        '2017-09-12',
        '2020-06-30', '2019-08-20', '2021-02-11', '2020-12-01', '2016-04-25',
        '2019-11-15',
        '2020-03-03', '2022-04-10', '2018-07-18', '2017-10-22', '2015-05-09',
        '2020-09-14',
        '2021-01-19', '2016-12-30', '2023-01-11', '2018-02-27', '2019-06-06',
        '2020-08-08',
        '2021-03-22', '2017-01-17'
    ]
}
```

```python
#convert dictionary to DataFrame
df=pd.DataFrame(data)
#display the DataFrame
print(df)
```

```python
df1=df[(df['Join_Date']>"2020-01-01")]
```

```python
print(df1)
```

```python
#calculate the average salary for each department
average_salary=df.groupby('Department')['Salary'].mean()
```

```python
print(average_salary)
```

```python
#Identify the department with the highest average salary
highest_avg_salary_dept=average_salary.idxmax()
```

```python
print(f"\nDepartment with the highest average_salary:{highest_avg_salary_dept}")
```

```python
#Add 'Years_With_Company' column
today = pd.to_datetime('today')
df['Years_with_Company'] = (today.year -pd.to_datetime(df['Join_Date']).dt.year)
```

```python
print(df)
```

```python
from datetime import date
df['Join_Date']=pd.to_datetime(df['Join_Date'])
today = date.today()
df['total_years_worked']=(pd.to_datetime(today)-df['Join_Date']).dt.days / 365
df
```

```python
#5. Sort by 'Years_with_Company' in descending order
df_sorted = df.sort_values('Years_with_Company', ascending=False)
print("\nEmployees sorted by years with company (descending):")
print(df_sorted)
```

```python
%matplotlib inline
```

```python
df_sorted.plot(y='Salary',x='Join_Date',kind='bar')
```
